#### Altair plots

```python
# Altair Year Sliders

period_selector_right = alt.selection_point(
    name="PeriodUpper",
    fields=["cutoff"],
    bind=alt.binding_range(min=min_year, max=max_year, step=1),
    value=[{"cutoff": max_year}], 
)

period_selector_left = alt.selection_point(
    name="PeriodLower",
    fields=["cutoff"],
    bind = alt.binding_range(min=min_year, max=max_year, step=1),
    value=[{"cutoff": min_year}],
)

# Altair Origin clickable legend (use SHIFT + Click to select multiple Origins)

origin_selection = alt.selection_point(fields=["Origin"])

origin_legend = (
    alt.Chart(pd.DataFrame(data={"Origin": ['Europe', 'USA', 'Japan']}))
    .mark_circle(size=300)
    .encode(x=alt.X("Origin:N"), color=alt.Color("Origin:N", legend=None), opacity=alt.condition(origin_selection, alt.value(1.0), alt.value(0.1)))
    .add_params(origin_selection)
)
```

```python
base = (
    alt.Chart(df)
    .properties(height=300)
    .transform_filter((alt.datum.YYYY <= period_selector_right.cutoff) & (alt.datum.YYYY >= period_selector_left.cutoff))
    .transform_filter(origin_selection)
    .add_params(period_selector_left, period_selector_right, origin_selection)
)
```

```python
bar_alt = (
    base.mark_bar().encode(
    x=alt.X('count(Origin):Q', title='Number of Records'), 
    y=alt.Y('Origin:N', title='Origin'), 
    color=alt.Color('Origin:N', legend=None))
)
```

```python
(bar_alt & origin_legend).display()
```
